```python
import sys
import subprocess

def ensure_package(pkg_name, upgrade=False):
    try:
        __import__(pkg_name)
        print(f"{pkg_name} 已安装")
    except ImportError:
        print(f"未检测到 {pkg_name}，正在安装...")
        cmd = [sys.executable, "-m", "pip", "install", pkg_name]
        if upgrade:
            cmd.append("--upgrade")
        subprocess.check_call(cmd)
        print(f"{pkg_name} 安装完成，请重启Jupyter kernel（点击Restart Kernel），否则新包无法立即导入。")

ensure_package("pygwalker", upgrade=False)
ensure_package("streamlit", upgrade=False)
```

```python
from pygwalker.api.streamlit import StreamlitRenderer
import pandas as pd
import streamlit as st

# Adjust the width of the Streamlit page
st.set_page_config(
    page_title="Use Pygwalker In Streamlit",
    layout="wide"
)

# Add Title
st.title("Use Pygwalker In Streamlit")

# You should cache your pygwalker renderer, if you don't want your memory to explode
@st.cache_resource
def get_pyg_renderer() -> "StreamlitRenderer":
    df = pd.read_csv(r"E:\kabuda_data_analysis\司机业务信息库\司机业务.csv")

    # If you want to use feature of saving chart config, set `spec_io_mode="rw"`
    return StreamlitRenderer(df, spec="./gw_config.json", spec_io_mode="rw")


renderer = get_pyg_renderer()

renderer.explorer()
```
